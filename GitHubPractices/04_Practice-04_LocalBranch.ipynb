## Process

Here's the step-by-step process of how I followed the course:

1. **git clone https://github.com/.../Repository.git**
    - clones the repository into local pc
    - I didn't exactly clone the repository but included for future reference

2. **mkdir GIT-Sample-Repo**
    - creates new directory (folder)

3. **cd GIT-Sample-Repo**
    - goes to the folder

4. **git branch PATCH-1-ADDING-HEADERS**
    - creates a local branch (but still working on the main branch)

5. **git checkout PATCH-1-ADDING-HEADERS**
    - switching/working inside the branch

6. **git status**
    - says which branch it's currently working on
    - lists changes (in existing files -- use git add) not stage for commit
    - lists all changes added (new files -- use git add) to commit

7. **git commit -m "Added header"**
    - commits changes inside the branch with the message written

8. **git checkout main**
    - changes to main branch to later merge

9. **git merge PATCH-1-ADDING-HEADERS**
    - merges changes commited inside branch PATCH-1-ADDING-HEADERS to the main branch (local)

10. **git push -u origin main**
    - pushes local main branch into the remote main branch

11. **git status**
    - checks if everything is pushed to the remote repository and that the local repository is up to date (no pull needed)
